```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = [10, 7]
```

```python
xaxis = np.loadtxt('values.txt')
yaxis = np.loadtxt('percentageinterval.txt')
txaxis = np.loadtxt('totalvalues.txt')
tyaxis = np.loadtxt('totalpercentageinterval.txt')
```

```python
plt.figure(1)
plt.plot(xaxis, yaxis, 'x', label = 'Chirp mass')
plt.plot(txaxis, tyaxis, 'x', color = 'red', label = 'Total mass')
plt.xlabel('Mass /M$_{sun}$')
plt.ylabel('Ratio of credible interval to mass')
plt.grid()
plt.legend(loc = 'best')
plt.savefig('ChirpVTotal')
```
